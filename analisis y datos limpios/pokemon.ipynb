```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np          
import nashpy as nash    
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("pokemon_limpio", sep=',') 
```

```python
df
```

```python




# Muestra la lista de columnas disponibles
print("Columnas en tu DataFrame:")
print(df.columns.tolist())

# Muestra las primeras filas para verificar los datos
print("\nPrimeras filas del DataFrame:")
print(df.head())
```

```python
# Preprocesamiento
le = LabelEncoder()
y = le.fit_transform(df['type1'])  # Convertimos tipos a números

# Características: todas las stats numéricas
features = ['hp', 'atk', 'def', 'spatk', 'spdef', 'speed', 'total']
X = df[features]

# Dividir datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Modelo
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)

# Evaluación
print(f"Precisión: {model.score(X_test, y_test):.2f}")
print(f"Tipos únicos: {le.classes_}")  # Mapeo de tipos
```

```python
from sklearn.linear_model import LinearRegression

X = df[['hp', 'atk', 'def', 'spatk', 'spdef', 'speed']]
y = df['total']

model = LinearRegression()
model.fit(X, y)

print(f"R²: {model.score(X, y):.2f}")  # Bondad del ajuste
```
