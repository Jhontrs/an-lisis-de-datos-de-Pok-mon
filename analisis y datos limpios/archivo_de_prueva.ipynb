```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración común
def plot_categorical(col, df, max_categories=20):
    plt.figure(figsize=(10, 5))
    sns.countplot(
        data=df,
        y=col,
        order=df[col].value_counts().iloc[:max_categories].index,
        palette="viridis"
    )
    plt.title(f'Distribución de {col} (n={df[col].nunique()})')
    plt.tight_layout()
    plt.show()

# Ejemplo para 'generation' (3 valores)
plot_categorical('generation', df)
```

```python
# Configuración común para gráficos
def plot_categorical(col, df, palette="viridis"):
    plt.figure(figsize=(10, 5))
    sns.countplot(
        data=df,
        y=col,
        order=df[col].value_counts().index,
        palette=palette
    )
    plt.title(f'Distribución de {col} (n={df[col].nunique()})')
    plt.tight_layout()
    plt.show()

# Ejemplo para 'generation' (3 valores)
plot_categorical('generation', df, "rocket")
```

```python
# Gráfico de tipos primarios vs. secundarios
type_counts = pd.concat([
    df['type1'].value_counts().rename('Primario'),
    df['type2'].value_counts().rename('Secundario')
], axis=1).fillna(0)

type_counts.plot(kind='bar', stacked=True, figsize=(12, 6), colormap="viridis")
plt.title('Tipos Primarios vs. Secundarios')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Agrupar categorías raras en "Otros"
def group_and_plot(df, col, top_n=30):
    top_categories = df[col].value_counts().head(top_n).index
    df[col + '_grouped'] = df[col].where(df[col].isin(top_categories), 'Otros')
    plot_categorical(col + '_grouped', df, "mako")

# Ejemplo para 'abilities' (top 30)
group_and_plot(df, 'abilities')
```

```python

df.columns = df.columns.str.strip()  # Limpiar espacios en nombres de columnas

# 2. Verificar valores únicos de generación
print("Valores únicos en columna 'generation':")
print(df['generation'].unique())

# 3. Filtrar SOLO PRIMERA GENERACIÓN (generation-i)
df_gen1 = df[df['generation'] == 'generation-i'].copy()
print(f"\n📊 Pokémon en 1ª generación: {len(df_gen1)}")

# 4. Definir stats con TUS nombres de columnas
stats = {
    'HP': 'hp',
    'Ataque': 'atk',
    'Defensa': 'def',
    'Ataque Especial': 'spatk',
    'Defensa Especial': 'spdef', 
    'Velocidad': 'speed'
}

# 5. Función para mostrar top 10
def mostrar_top_10(df, stats_dict):
    for stat_name, stat_col in stats_dict.items():
        if stat_col in df.columns:
            top_10 = df.nlargest(10, stat_col)[['name', stat_col]]
            print(f"\n🔝 Top 10 en {stat_name}:")
            print(top_10.to_string(index=False))
            
            # Gráfico
            plt.figure(figsize=(10, 5))
            sns.barplot(data=top_10, y='name', x=stat_col, palette='viridis')
            plt.title(f'Top 10 {stat_name} (Gen 1)')
            plt.show()
        else:
            print(f"\n⚠️ Columna '{stat_col}' no encontrada")

# 6. Mostrar resultados
if len(df_gen1) > 0:
    print("\n⭐ ANÁLISIS 1ª GENERACIÓN (KANTO) ⭐")
    mostrar_top_10(df_gen1, stats)
    
    # Stats promedio
    print("\n📈 STATS PROMEDIO:")
    print(df_gen1[['hp', 'atk', 'def', 'spatk', 'spdef', 'speed']].mean().round(1))
else:
    print("\n❌ No se encontraron Pokémon de 1ª generación")
    print("Revisa que 'generation-i' exista en tus datos")
```

```python

df.columns = df.columns.str.strip()
df_gen1 = df[df['generation'] == 'generation-i'].copy()

# 2. Configuración de estilos
sns.set_style("whitegrid")
plt.figure(figsize=(15, 10))

# 3. Gráfico de distribución para cada stat
stats = ['hp', 'atk', 'def', 'spatk', 'spdef', 'speed']
titles = ['HP', 'Ataque', 'Defensa', 'Ataque Especial', 'Defensa Especial', 'Velocidad']

for i, (stat, title) in enumerate(zip(stats, titles), 1):
    plt.subplot(2, 3, i)
    sns.histplot(df_gen1[stat], kde=True, color='skyblue', bins=15)
    plt.title(f'Distribución de {title}')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    
    # Añadir líneas de media y mediana
    mean_val = df_gen1[stat].mean()
    median_val = df_gen1[stat].median()
    plt.axvline(mean_val, color='red', linestyle='--', label=f'Media: {mean_val:.1f}')
    plt.axvline(median_val, color='green', linestyle=':', label=f'Mediana: {median_val:.1f}')
    plt.legend()

plt.tight_layout()
plt.suptitle('Distribuciones de Stats - Primera Generación (Kanto)', y=1.02, fontsize=16)
plt.show()

# 4. Boxplot comparativo
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_gen1[stats], palette="Set2")
plt.title('Comparación de Stats - Primera Generación')
plt.xticks(ticks=range(len(stats)), labels=titles)
plt.ylabel('Valor')
plt.show()

# 5. Matriz de correlación
plt.figure(figsize=(10, 8))
corr_matrix = df_gen1[stats].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlación entre Stats\nPrimera Generación')
plt.show()
```

```python

df_gen1 = df[df['generation'] == 'generation-i'].copy()

# 2. Configuración de estilos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)  # Tamaño único para todos los gráficos

# 3. Diccionario de stats (usando tus nombres de columnas)
stats = {
    'HP': 'hp',
    'Ataque': 'atk',
    'Defensa': 'def',
    'Ataque Especial': 'spatk',
    'Defensa Especial': 'spdef',
    'Velocidad': 'speed'
}

# 4. Generar gráficos interactivos
for stat_name, stat_col in stats.items():
    # Crear figura nueva para cada stat
    plt.figure()
    
    # Histograma + KDE
    ax = sns.histplot(
        data=df_gen1,
        x=stat_col,
        kde=True,
        bins=15,
        color='skyblue',
        edgecolor='white'
    )
    
    # Líneas de media/mediana
    mean_val = df_gen1[stat_col].mean()
    median_val = df_gen1[stat_col].median()
    
    plt.axvline(mean_val, color='red', linestyle='--', label=f'Media: {mean_val:.1f}')
    plt.axvline(median_val, color='green', linestyle=':', label=f'Mediana: {median_val:.1f}')
    
    # Personalización
    plt.title(f'Distribución de {stat_name}\nPrimera Generación (Kanto)', fontsize=14)
    plt.xlabel('Valor del Stat', fontsize=12)
    plt.ylabel('Frecuencia', fontsize=12)
    plt.legend()
    
    # Mostrar gráfico (uno por uno)
    plt.tight_layout()
    plt.show()

# Mensaje final
print("🎉 Visualización completada. Cierra cada gráfico para continuar.")
```
